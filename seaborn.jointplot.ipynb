```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("attention")
```

```python
data.head()
```

```python
sns.jointplot(x = "solutions", y = "score",
              kind = "hex", data = data)
plt.show()
```

```python
data = sns.load_dataset("mpg")
```

```python
data.head()
```

```python
sns.jointplot(x = "mpg", y = "acceleration",
              kind = "scatter", data = data)
plt.show()
```
